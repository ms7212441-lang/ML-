````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

iris = datasets.load_iris()

X = iris.data[:, :2]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model = SVC(kernel='linear')

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

print("\nTest Data with Predictions:")
for i in range(len(X_test)):
    print("Input:", X_test[i], " Predicted:", y_pred[i], " Actual:", y_test[i])

def plot_decision_boundary(X, y, model):
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, h),
        np.arange(y_min, y_max, h)
    )

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("SVM Decision Boundary")
    plt.show()

plot_decision_boundary(X, y, model)
````

### Algorithm for SVM Classification and Decision Boundary Visualization

1.  **Import Libraries**: Import `numpy`, `pandas`, `matplotlib.pyplot`, `datasets`, `train_test_split`, `SVC`, `accuracy_score`, `confusion_matrix`, and `classification_report` from `sklearn`.
2.  **Load Data**: Load the Iris dataset.
3.  **Prepare Features and Target**: Select the first two features (sepal length and sepal width) as `X` and the target variable as `y`.
4.  **Split Data**: Split the dataset into training and testing sets (`X_train`, `X_test`, `y_train`, `y_test`) with a test size of 30% and a fixed `random_state`.
5.  **Initialize SVM Model**: Create an `SVC` (Support Vector Classifier) model with a linear kernel.
6.  **Train Model**: Fit the SVM model to the training data (`X_train`, `y_train`).
7.  **Make Predictions**: Use the trained model to predict the target values for the test data (`X_test`).
8.  **Evaluate Model**: Calculate and print the accuracy score, confusion matrix, and classification report based on the true test labels (`y_test`) and predicted labels (`y_pred`).
9.  **Print Test Data with Predictions**: Iterate through the test data and print each input, its predicted class, and its actual class.
10. **Define `plot_decision_boundary` Function**:
    a.  Set the step size `h` for the meshgrid.
    b.  Determine the min and max values for the features `X[:, 0]` and `X[:, 1]` to define the plot boundaries.
    c.  Create a meshgrid (`xx`, `yy`) over the feature space.
    d.  Predict the class for each point in the meshgrid using the trained model.
    e.  Reshape the predictions to match the meshgrid shape.
    f.  Plot the decision boundary using `plt.contourf`.
    g.  Plot the actual data points using `plt.scatter`, colored by their true labels.
    h.  Add labels for the x and y axes and a title to the plot.
    i.  Display the plot.
11. **Call `plot_decision_boundary`**: Call the defined function with the full dataset (`X`, `y`) and the trained model to visualize the decision boundary.
