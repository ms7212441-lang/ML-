````
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('/content/ml 8.csv')

print("Dataset:\n")
print(data)

X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

weights = np.ones(X.shape[1])
threshold = X.shape[1] / 2

def mc_pitts_neuron(inputs, weights, threshold):
    net_input = np.dot(inputs, weights)
    return 1 if net_input >= threshold else 0

predictions = []
for sample in X:
    output = mc_pitts_neuron(sample, weights, threshold)
    predictions.append(output)

0 / 100000

predictions = np.array(predictions)

if X.shape[1] == 2:
    plt.figure(figsize=(6,5))

    plt.scatter(X[:, 0], X[:, 1], c=y, marker='o', label='Actual')
    plt.scatter(X[:, 0], X[:, 1], c=predictions, marker='x', label='Predicted')

    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("McCulloch-Pitts Model Output")
    plt.legend()

    plt.grid(True)
    plt.show()

else:
    plt.figure(figsize=(6,5))

    plt.plot(y, label='Actual', marker='o')
    plt.plot(predictions, label='Predicted', marker='x')

    plt.title("Actual vs Predicted")
    plt.legend()

    plt.grid(True)
    plt.show()


accuracy = np.mean(predictions == y)
print("\nAccuracy:", accuracy)
````

### Algorithm: McCulloch-Pitts Neuron Model

1. **Data Acquisition**: Load the dataset from a CSV file using pandas.
2. **Preprocessing**:
   - Separate features ($X$) and target labels ($y$).
   - Convert the data into NumPy arrays for efficient computation.
3. **Parameter Initialization**:
   - Initialize weights ($w$) as a vector of ones with the same length as the number of features.
   - Set the threshold ($	heta$) to half the number of features.
4. **Neuron Function Definition**:
   - Define `mc_pitts_neuron(inputs, weights, threshold)`:
     - Calculate the net input ($Z$) by computing the dot product of `inputs` and `weights` ($Z = \sum_{i=1}^{n} x_i w_i$).
     - Apply the threshold logic: If $Z \geq \theta$, return 1; otherwise, return 0.
5. **Model Prediction**:
   - Iterate through each sample in the dataset $X$.
   - Pass each sample through the `mc_pitts_neuron` function.
   - Store the results in a list called `predictions`.
6. **Visualization**:
   - If the data is 2D, create a scatter plot comparing actual vs. predicted labels.
   - If the data is not 2D, use a line plot to show the comparison.
7. **Evaluation**:
   - Compare the `predictions` array with the actual target labels $y$.
   - Calculate accuracy as the mean of correct predictions.

````

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/content/ml 9.csv')


X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
print("Dataset:")
print(data)

weights = np.zeros(X.shape[1])
bias = 0
learning_rate = 0.1
epochs = 10


def step_function(x):
    return 1 if x >= 0 else 0


for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        y_pred = step_function(linear_output)


        error = y[i] - y_pred
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Sample {X[i]}, Predicted: {y_pred}, Error: {error}, Weights: {weights}, Bias: {bias}")


print("\nTesting trained perceptron:")
predictions = []
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    y_pred = step_function(linear_output)
    predictions.append(y_pred)
    print(f"Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}")

predictions = np.array(predictions)

if X.shape[1] == 2:
    plt.figure(figsize=(6,5))

    plt.scatter(X[:,0], X[:,1], c=y, marker='o', label='Actual')

    plt.scatter(X[:,0], X[:,1], c=predictions, marker='x', label='Predicted')


    x1_vals = np.linspace(np.min(X[:,0])-0.1, np.max(X[:,0])+0.1, 100)
    if weights[1] != 0:
        x2_vals = -(weights[0]*x1_vals + bias)/weights[1]
        plt.plot(x1_vals, x2_vals, 'r--', label='Decision Boundary')

    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("Single-Layer Perceptron Decision Boundary")
    plt.legend()
    plt.grid(True)
    plt.show()
else:
    print("Plotting only works for 2D feature data.")
````

### Algorithm: Single-Layer Perceptron Learning

1. **Data Preparation**: Load the dataset and separate the input features ($X$) and target labels ($y$).
2. **Initialization**:
   - Set weights ($w$) to zeros.
   - Set bias ($b$) to 0.
   - Define the learning rate ($η$) and the number of epochs.
3. **Activation Function**: Define a step function that returns 1 for inputs $\ge 0$ and 0 otherwise.
4. **Training (Iterative Learning)**:
   - For each epoch:
     - For each sample $x_i$ in the dataset:
       - Calculate the linear output: $Z = (X_i \cdot w) + b$.
       - Pass $Z$ through the step function to get the prediction $\hat{y}$.
       - Calculate the error: $e = y_i - \hat{y}$.
       - Update weights: $w = w + η \cdot e \cdot X_i$.
       - Update bias: $b = b + η \cdot e$.
5. **Testing**: Use the final weights and bias to predict labels for the dataset.
6. **Visualization**: Plot the data points and the derived decision boundary line defined by $w_1x_1 + w_2x_2 + b = 0$.

### Pseudocode: Single-Layer Perceptron

```text
BEGIN
    LOAD data FROM "ml 9.csv"
    SET X = features, y = labels
    
    INITIALIZE weights = [0, 0], bias = 0
    SET learning_rate = 0.1, epochs = 10
    
    FUNCTION step_function(x):
        RETURN 1 IF x >= 0 ELSE 0

    FOR epoch FROM 1 TO epochs:
        FOR each sample X[i] with target y[i]:
            linear_output = dot_product(X[i], weights) + bias
            prediction = step_function(linear_output)
            
            error = y[i] - prediction
            
            weights = weights + (learning_rate * error * X[i])
            bias = bias + (learning_rate * error)
        END FOR
    END FOR

    FOR each sample X[i]:
        final_prediction = step_function(dot_product(X[i], weights) + bias)
        DISPLAY final_prediction
    END FOR
    
    IF features == 2:
        PLOT decision_boundary: weights[0]*x1 + weights[1]*x2 + bias = 0
END
```
